```r
#Using cluster information from :
# awk  '{print $3"\t"$9}' GSE129785_scATAC-Hematopoiesis-All.cell_barcodes.txt  > cluster_names
get_our_cluster <- function(cluster_name){
    i <- strsplit(cluster_name,split='Cluster')[[1]][2]
    if (i %in% 1:9){
       return('Cluster1')
   }
    else if (i %in% 10:11){
       return('Cluster2')
   }
    else if (i %in% 12:13){
       return('Cluster3')
   }
    else if (i %in% 14:16){
       return('Cluster4')
   }
    else if (i %in% 17:17){
       return('Cluster5')
   }
    else if (i %in% 18:20){
       return('Cluster6')
   }
    else if (i %in% 21:25){
       return('Cluster7')
   }
    else if (i %in% 26:31){
       return('Cluster8')
   }  
    else {
        return("NA")
    }
}
clusters <- read.table("cluster_names",sep="\t", header=TRUE,comment.char="")

df1 <- clusters
finalcells <- projHeme1$cellNames
clusters <- df1$Clusters[match(finalcells,df1$Group_Barcode)]
our_clusters <- c()
for (c in clusters){
    our_clusters <- append(our_clusters, get_our_cluster(c))
}
```

```r
projHeme1 <- addCellColData(ArchRProj = projHeme1, data = our_clusters,
      cells = finalcells, name = "Cluster")
```

```r
projHeme1 <- addGroupCoverages(
  ArchRProj = projHeme1,
  groupBy = "Cluster",
  minCells=10,
  maxCells = 4000,  #monocyte cluster has 4022 cells in PBMC_Rep1 group
  maxReplicates = 16, # monocyte cluster has 16 groups
  force =TRUE
)
```

```r
# sanity checks
# 1) heatmap of cell subclass by experiment_short
install.packages('pheatmap')
library(pheatmap)

#unnomralized_sample_heatmap
cM <- confusionMatrix(paste0(projHeme1$Cluster), paste0(projHeme1$Sample))
pheatmap::pheatmap( mat = as.matrix(t(cM)), color = paletteContinuous("whiteBlue"), border_color = "black")

# nomralized_sample_heatmap
cM <- cM / Matrix::rowSums(cM)
pheatmap::pheatmap( mat = as.matrix(t(cM)), color = paletteContinuous("whiteBlue"), border_color = "black")


# I guess these heatmaps are as expected because peaks haven't been called yet. 
#These are fragments from Satpathy group who have already shown in their analysis that their peaks are good. 
```

```r
pathToMacs2 <- findMacs2()
projHeme1 <- addReproduciblePeakSet(
    ArchRProj = projHeme1,
    groupBy = "Cluster",
    pathToMacs2 = pathToMacs2,
    force=TRUE
)
```

```r
saveArchRProject(ArchRProj = projHeme1, outputDirectory = "Save-ProjHeme1", load = FALSE)
```

```r
library(dplyr)
gr <- getPeakSet(projHeme1)
head(gr)
df <- data.frame(seqnames=seqnames(gr),
  starts=start(gr),
  ends=end(gr),
  names=names(gr),
  scores=score(gr),
  strands=strand(gr))
```

```r
for (i in c(1,2,3,4,5,6,7,8)){
    clus <- paste('Cluster',i,sep="")
    sub_df <- subset(df,names==clus)
    write.table(sub_df, file=paste(clus,".bed",sep=""), quote=F, sep="\t", row.names=F, col.names=F)
}
```
